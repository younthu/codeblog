```python
from jinja2 import Template,PackageLoader,Environment,FileSystemLoader

navigationIndex = 8
supportedByMobileIndex = 5
moduleIndex  = 2
useCaseIndex = 3
userStoryIndex = 4
ids = []
Navi = namedtuple('Navi', ['action','entity', 'module', 'navigationId', 'useCase', 'userStory', 'upperCaseEnum'])
for _, row in df.iterrows():
#     print(row)
    supportedByMobile = row[supportedByMobileIndex] 
    navigationId=row[navigationIndex]
    if pd.isnull(supportedByMobile) or not supportedByMobile.startswith('Y') or not navigationId:
        continue
    action = navigationId.split('/')[-1]
    entity = navigationId.split('/')[-2]
    
    enumName = (action + entity).replace(".", "_")
    upperCaseEnum = "_".join(l.lower() for l in re.findall('[A-Z][^A-Z]*', enumName)).upper()

    module = row[moduleIndex]
    useCase = row[useCaseIndex]
    userStory = row[userStoryIndex]

    ids.append(Navi(action=action, navigationId=navigationId, entity=entity, userStory=userStory, 
                    module=module, useCase=useCase, upperCaseEnum=upperCaseEnum))
    print(navigationId)
    
env = Environment(loader = FileSystemLoader("./"))
template = env.get_template("SFNavigationID.swift")  

content = template.render(ids=ids)
print(content)
```
